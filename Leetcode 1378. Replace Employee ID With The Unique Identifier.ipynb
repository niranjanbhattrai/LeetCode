#### 1378. Replace Employee ID With The Unique Identifier

```python
from pyspark.sql import SparkSession, Row
import pyspark.sql.functions as f
import pandas as pd
import numpy as np
from pyspark import SparkContext
sc = SparkContext()
sc.setLogLevel("ERROR")
spark = SparkSession.builder\
    .master("local")\
    .appName("main")\
    .config("spark.dynamicAllocation.enabled", "true")\
    .config("spark.shuffle.service.enabled", "true")\
    .config("spark.sql.catalogImplementation", "hive")\
    .getOrCreate()
```

#### Solution using Pandas

```python
data = [[1, 'Alice'], [7, 'Bob'], [11, 'Meir'], [90, 'Winston'], [3, 'Jonathan']]
employees = pd.DataFrame(data, columns=['id', 'name']).astype({'id':'int64', 'name':'object'})
data = [[3, 1], [11, 2], [90, 3]]
employee_uni = pd.DataFrame(data, columns=['id', 'unique_id']).astype({'id':'int64', 'unique_id':'int64'})
```

```python
employees.merge(employee_uni, on= "id", how= "left")[["unique_id", "name"]]
```

#### Solution using Pyspark

```python
employees = spark.createDataFrame(employees)
employee_uni = spark.createDataFrame(employee_uni)
```

```python
employees.join(employee_uni, on = (employees.id == employee_uni.id), how= "left").select("unique_id", "name").show()
```
